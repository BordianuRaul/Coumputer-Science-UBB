```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.tree import plot_tree
from sklearn import tree

from sklearn.model_selection import GridSearchCV 


df_iris = pd.read_csv('iris_teach_2.csv')

empty_cells = df_iris.isnull().sum()

df_iris.dropna(inplace=True)

X = df_iris.iloc[:, :-1]

Y = df_iris.iloc[:, -1]

label_encoder = LabelEncoder()

label_encoder.fit(Y)

y_encoded = label_encoder.transform(Y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

X_modified = X.drop(columns=['petal width (cm)'])

tree_clf = DecisionTreeClassifier(criterion='entropy', max_depth=2,random_state=42)

tree_clf.fit(X_modified, y_encoded)

X_test_modified = X_test.drop(columns=['petal width (cm)'])
y_pred = tree_clf.predict(X_test_modified)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}".format(accuracy))
ram_grid = {"max_depth": [1, 2, 3, 4, 5, 6, 7]} 

# Create a grid search object
grid_search = GridSearchCV(tree_clf, ram_grid, cv=5) 

# Fit the grid search object to the training data
grid_search.fit(X_modified, y_encoded) 

# Print the best hyperparameters found by the grid search
print("Best hyperparameters:", grid_search.best_params_)

# Get the best model with the optimal depth
best_tree_clf = grid_search.best_estimator_

# Evaluate the score of the best model
best_score = best_tree_clf.score(X_test_modified, y_test)
print("Accuracy with the best model: {:.2f}".format(best_score))

print("Textual representation of the decision tree:")
text_representation = tree.export_text(best_tree_clf)
print(text_representation)

# Visualize the decision tree
plot_tree(best_tree_clf)
```
